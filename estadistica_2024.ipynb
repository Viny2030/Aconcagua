 # **<font color="DeepPink">Introducci√≥n</font>**

 <p align="justify">
Podr√≠amos decir que el <strong>problema general de la estad√≠stica</strong> es determinar como se distribuye una V.A. $X$ cualquiera. Es decir, estamos interesados en obtener informaci√≥n relativa una V.A. que sea relevante para la toma de decisiones.<br><br>
üëÄ Ejemplo:<br><br>
Estoy interesado en conocer <strong>c√≥mo me puede ir en t√©rminos salariales si me mudo a Espa√±a</strong>. Necesito informaci√≥n sobre como se distribuyen los salarios en Espa√±a, mejor si es informaci√≥n relativa a como se distribuyen los salarios en la regi√≥n que estoy pensando mudarme, y mejor a√∫n si la informaci√≥n es relativa a salarios de la regi√≥n vinculados al nivel educativo que tengo y puesto laboral al que puedo aspirar.<br><br>
Mientras m√°s informaci√≥n vinculada con las condiciones que yo necesito tenga, m√°s preciso tender√° a ser mi an√°lisis y m√°s certera tender√° a ser la decisi√≥n que tome.<br><br>
La informaci√≥n necesaria para intentar reconstruir la distribuci√≥n de prababilidad de una V.A. provendr√° de una muestra aleatoria.<br><br>
‚úÖ <strong>Poblaci√≥n</strong>: Se llama poblaci√≥n al conjunto total de
elementos en discusi√≥n y sobre los cu√°les se quiere tener alguna informaci√≥n.<br><br>
‚úÖ <strong>Muestra aleatoria</strong>: Una muestra aleatoria de tama√±o $n$
de una poblaci√≥n con densidad $f_X$ es un conjunto de $n$ variables aleatorias independientes y cada una con id√©ntica distribuci√≥n en la poblaci√≥n.
Se denota:<br><br>
$$ X_1,X_2,...,X_n \sim^{i.i.d} f_{X} $$<br><br>
En general, obtenemos muestras de <strong>vectores aleatorios</strong> cada uno formado por un conjunto de V.A. que esperamos est√©n relacionadas de alguna forma, pero cada vector independiente de los restantes. Con la muestra de estos vectores aleatorios queremos:<br><br>
1. Determinar la distribuci√≥n de $X$ condicionada a cada variable relevante para explicarala. Queremos el v√≠nculo entre las variables. Este es el escenario √≥ptimo pero no es posible.<br><br>
2. Determinar la distribuci√≥n de $X$ condicionada a algunas variables relevantes para explicarala (o algunas caracter√≠sticas de esta distribuci√≥n). Veremos que es muy costoso en t√©rminos de informaci√≥n.<br><br>
3. Determinar la distribuci√≥n $X$ condicionada a algunas variables relevantes para explicarala suponiendo alguna relaci√≥n a priori entre ellas. Menos costoso pero m√°s impreciso.<br><br>
4. Determinar la distribuci√≥n $X$ no condicionada o algunas caracter√≠sticas de  esta distribuci√≥n. Sigue siendo muy costoso y complejo de interpretar.<br><br>
</p>

 # **<font color="DeepPink">Estimador</font>**

 <p align="justify">
‚úÖ <strong>Estad√≠stico</strong><br><br>
Dada una muestra aleatoria $X_1, X_2, ..., X_n$, un estad√≠stico es una funci√≥n aplicada a la muestra aleatoria cuyo resultado es una nueva variable aleatoria. Por ejemplo los estad√≠sticos $\bar{X}$ o $\tilde{S^2}$ vienen dados por:<br><br>
$$ \bar{X} = \frac{1}{n}\sum_{i=1}^{n}X_i $$<br><br>
$$ \tilde{S^2} = \frac{1}{n}\sum_{i=1}^{n}(X_i - \bar{X})^2 $$<br><br>
A $\bar{X}$ se lo denomina promedio o media muestral y a $\tilde{S^2}$ cuasi-varianza muestral.<br><br>
Dado que los estadisticos son V.A., <strong>tienen su propia distribuci√≥n con sus propias carcater√≠sticas</strong>, las cuales derivan de la distribuci√≥n original de $X$. Ejemplo:<br><br>
$$ E(\bar{X}) = E(X) $$<br><br>
Esto nos dice que "en promedio", la media muestral es igual a la poblacional. Es decir, a veces la media muestral toma valores por debajo de la poblacional y a veces por encima pero de una forma "equilibrada".<br><br>
Otra caracteristica es:<br><br>
$$ \text{var}(\bar{X}) = \frac{\text{var}(X)}{n} $$<br><br>
La varianza de $\bar{X}$ es igual a la varianza de $X$ partida por el tama√±o de la muestra. Mientras mayor sea la muestra, menor ser√° la variabilidad de $\bar{X}$. De esto se deduce:<br><br>
$$ \sigma_{\bar{X}}=\frac{\sigma_X}{\sqrt{n}}  $$<br><br>
‚úÖ <strong>Estimador</strong><br><br> Un estimador es un estadistico cuyo valor observado se utiliza para estimar alguna carcater√≠stica o par√°metro de la distribuci√≥n de una Variable Aleatoria.<br><br>
Puesto que $\bar{X}$ en promedio toma el valor de la verdadera $E(X)$ es intuitivo considerar que el valor observado de $\bar{X}$ nos puede servir para estimar a $E(X)$.<br><br>
‚úÖ <strong>¬øCu√°ndo un estimador es bueno?</strong><br><br>
Dado un estadistico $\hat{\theta}$, este estad√≠stico ser√° un buen estimador de un par√°metro $\theta$ de una distribuci√≥n dada si la <strong>distancia</strong> entre el estadistico y el verdadero par√°metro es peque√±a. A esta distancia se la conoce como <strong>Error Cuadr√°tico Medio</strong>.<br><br>
$$ ECM(\hat{\theta})=E\big[ (\hat{\theta} - \theta)^2 \big] $$<br><br>
Cuando un estimador $\hat{\theta}$ cumple que $E(\hat{\theta})=\theta$ (como el caso de la media muestral) decimos que el estimador es insesgado. Podemos descomponer el Error Cuadratico Medio de la siguiente forma:<br><br>
$$ ECM(\hat{\theta}) = \text{var}(\hat{\theta}) + \big( E(\hat{\theta}) - \theta  \big)^2 $$<br><br>
El $ECM$ de un estimador se compone de la varianza del estimador m√°s el sesgo del estimador elevado al cuadrado. Para el caso de usar el promedio muestral para estimar la media poblacional tenemos que:<br><br>
$$ ECM(\bar{X}) = \text{var}(\bar{X}) + \big( E(\bar{X}) - E(X))  \big)^2 = \frac{\text{var}(X)}{n} $$<br><br>
El $ECM$ del promedio es la varianza de la variable aleatoria dividida por el tama√±o muestral. Mientras mayor sea la muestra mejor funciona el estimador.<br><br>
De aqu√≠ surgen 3 propiedades deseables de los estimadores:<br><br>
1. <strong>Insesgadez</strong>: $E(\hat{\theta})=\theta$.<br><br>
2. <strong>Varianza m√≠nima</strong> (tambi√©n eficiencia o precisi√≥n).<br><br>
3. <strong>Consistencia</strong>: $ \lim_{n \to \infty}ECM(\hat{\theta})=0$. Cuando el tama√±o de muestra aumenta indefinidamente el estimador converge al verdadero valor del par√°metro estimado.<br><br>
Como hemos visto, $\bar{X}$ es un estimador insesgado de $E(X)$. Adem√°s, cumple que:<br><br>
$$ \lim_{n \to \infty} \text{var}(\bar{X})=\lim_{n \to \infty} \frac{\text{var}(X)}{n} = 0 $$.<br><br>
Estas dos propiedades hacen que sea un estimador consistente.<br><br>
$ \tilde{S^2} $ es un estimador sesgado de $\text{var}(X)$ puesto que:<br><br>
$$ E(\tilde{S^2}) = \frac{n-1}{n}\sigma^2 $$.<br><br>
Para tama√±os muestrales muy grandes el sesgo tiende a desaparecer por lo que se dice que es <strong>asint√≥ticamente insesgado</strong>. Tambi√©n podemos corregir el sesgo usando el siguiente estimador para la varianza de una V.A. $X$:<br><br>
$$ S^2 = \frac{n}{n-1}\tilde{S^2} = \frac{1}{n-1} \sum_{i=1}^{n}(X_i - \bar{X})^2 $$.<br><br>
A $S^2$ lo denominamos varianza muestral.<br><br>
Existen otras propiedades importantes que se suelen pedir a los estimadores. Por ejemplo, la <strong>robustez</strong>, que implica que el estimador no es muy sensible a cambios en los valores que toma la muestra.





</p>

# **<font color="DeepPink">¬øC√≥mo obtenemos estimadores?</font>**

 <p align="justify">
A grandes razgos, hay dos m√©todos para obtener estimadores para una V.A. $X$.<br><br>
1. Por optimizaci√≥n (M-estimadores).<br><br>
2. Por construcci√≥n similar de lo que intentan estimar (estimadores de momentos).<br><br>
‚úÖ <strong>M-estimadores</strong><br><br>
Existen dos mecanimos para obtener M-estimadores. En el primero que vamos a estudiar, denominado m√©todo de m√°xima verosimilitud, se establece, a priori, una posible familia de distribuciones para $X$ (normal, beta, gamma, etc.) y se estiman los par√°metros de esa distribuci√≥n de forma tal que la funci√≥n de probabilidad resultante sea la que mejor "ajuste" los datos observados.<br><br>
Si consideramos que $X \sim f(\cdot, \theta)$, donde $\theta$ es el par√°metro (o vector de par√°metros) a estimar, dado que la muestra aleatoria genera variables independientes identicamente distribuidas, tenemos que la funci√≥n densidad de probabilidad conjunta de toda la muestra viene dada por:<br><br>
$$ f_{X_1,X_2,...,X_n}(x_1,x_2,...,x_n,\theta) = f_{X}(x_1,\theta)\times f_{X}(x_2,\theta)\times ...\times f_{X}(x_n,\theta) $$.<br><br>
Utilizamos los datos observados de la muestra para $x_1, x_2,...,x_n$ y buscamos $\hat{\theta}$ de forma tal que:<br><br>
$$ \hat{\theta} = \arg\max_{\theta} \Big\{ f_{X}(x_1,\theta)\times f_{X}(x_2,\theta)\times ...\times f_{X}(x_n,\theta) \Big\} $$.<br><br>
A la funci√≥n $f_{X}(x_1,\theta)\times f_{X}(x_2,\theta)\times ...\times f_{X}(x_n,\theta)$ se la conoce como <strong>funci√≥n de verosimilitud</strong>.<br><br>
En general, se aplica logaritmos (naturales) a la funci√≥n de verosimilitud para obtener una suma donde hay una multiplicaci√≥n de funciones. De esta forma, el proceso de derivaci√≥n (y optimizaci√≥n en general) es m√°s sensillo:<br><br>
$$ \hat{\theta} = \arg\max_{\theta} \Big\{ \log(f_{X}(x_1,\theta)) + \log(f_{X}(x_2,\theta)) + ... + \log( f_{X}(x_n,\theta)) \Big\} $$.<br><br>
A la nueva funci√≥n a optimizar se la conoce como log-verosimilitud y este proceso produce el mismo resultado que el anterior.<br><br>
El segundo mecanismo para obtener M-estimadores consiste en establecer una funci√≥n de p√©rdida que depende de la muestra y del par√°metro desconocido a estimar, y se utilizan los datos observados de la muestra para minimizar la funci√≥n de p√©rdida. Este m√©todo no requiere suponer como se distribuye la variable aleat√≥ria $X$ bajo estudio.<br><br>
Simulemos un ejemplo de una distribuci√≥n normal.



</p>

```python
#Construimos una muestra simulada de una distribuci√≥n normal con media 3 y varianza 4
#np.random.seed(12321)
muestra = sc.stats.norm.rvs(loc=3, scale=2, size=100)
```

```python
#Construimos la funci√≥n densidad de una normal dejando como variables sus par√°metros
def f(x,mu,sigma):
  return 1/(np.sqrt(2*(sigma**2)*np.pi)) * np.e**(-((x - mu)**2)/(2*sigma**2))
```

```python
#Construimos la funci√≥n de log-verosimilitud de la mustra, que solo depende de mu y sigma (los par√°metros a estimar)
def logV(param):
  v=[]
  for x in muestra:
    v.append(np.log(f(x,param[0],param[1])))
  return sum(v)
```

```python
logV([3,2])
```

```python
from scipy.optimize import minimize_scalar
#Esta funci√≥n est√° construida para minimizar, no para maximizar
```

La funci√≥n "minimize_scalar" busca el punto del dominio de $f$ donde presenta un m√≠nimo. Para resolver este problema tenemos el siguiente resultado.<br><br>
Resolver el problema de maximizar una cierta funci√≥n $f$ equivale a resolver el problema de minimizar la funci√≥n $-f$. Esto es, $f$ presenta un m√°ximo en $\theta_0$ si y solo si $-f$ presenta un m√≠nimo en $\theta_0$.

```python
#Construimos la opuesta de la funci√≥n de log-verosimilitud, funci√≥n que vamos a minimizar.
def menoslogV(param):
  v=[]
  for x in muestra:
    v.append(-np.log(f(x,param[0],param[1])))
  return sum(v)
```

```python
param0 = (1, 3)
param_hat = sc.optimize.minimize(menoslogV, param0, method='BFGS').x
param_hat
```

# **<font color="DeepPink">Ley de los grandes n√∫meros y el teorema central del l√≠mite</font>**

 <p align="justify">
Dos (o tres) resultados importantes:<br><br>
1. <strong> Ley de los grandes n√∫meros: </strong>Sean $X_1,..., X_n$ una muestra aleatoria de una poblaci√≥n con densidad desconocida pero con media $\mu$ finita, entonces:<br><br>
$$ \bar{X_n}  \xrightarrow{p} \mu \quad \text{a medida que} \quad n \to \infty  $$<br><br>
Intuitivamente, a medida que el tama√±o muestral crece el promedio se parace m√°s a la verdadera media.<br><br>
2. <strong> Teorema centr√°l del l√≠mite: </strong>Sean $X_1,..., X_n$ una muestra aleatoria de una poblaci√≥n con densidad desconocida pero con media $\mu$ y varianza $\sigma^2$ finitas, entonces:<br><br>
$$ Z_n = \frac{\bar{X_n} - \mu}{\frac{\sigma}{\sqrt{n}}} \xrightarrow{d} N(0,1) \quad \text{a medida que} \quad n \to \infty  $$<br><br>
Intuitivamente, a medida que el tama√±o muestral crece la distribuci√≥n de la variable aleatoria $Z$ (estandarizada) se parece m√°s y m√°s a una distribuci√≥n $N(0,1)$.<br><br>
3. <strong> M√©todo delta: </strong> sean $X_1,..., X_n$ una muestra aleatoria de una poblaci√≥n con densidad desconocida pero con media $\mu$ y varianza $\sigma^2$ finitas y $g$ una funci√≥n, entonces:<br><br>
$$ Z_n = \frac{g(\bar{X_n}) - g(\mu)}{\frac{\sigma \times g'(\sigma)}{\sqrt{n}} } \xrightarrow{d} N(0,1) \quad \text{a medida que} \quad n \to \infty  $$<br><br>
</p>

# **<font color="DeepPink">Estimaci√≥n por intervalos</font>**

 <p align="justify">
El principal problema con la metodolog√≠a anterior radica en que no sabemos que tan alejada est√° nuestra estimaci√≥n del verdadero par√°metro que estamos estimando. Se utilizan intervalos de confianza para intentar resolver este problema.<br><br>

Sea $X_1,X_2,...,X_n$ una muestra aleatoria de una poblaci√≥n con densidad $f_X(\cdot| \theta)$. Sean $T_1$ y $T_2$ dos estad√≠sticos que complen que $P(T_1 < T_2) = 1$ y para los cuales se cumple que:<br><br>
$$ P(T_1 < \theta < T_2) = \gamma $$<br><br>
Entonces $(T_1,T_2)$ es un intervalo del $\gamma$ por ciento de confianza para $\theta$.<br><br>
Algunos puntos importantes:<br><br>
* $\gamma$ se debe interpretar como la probabildad de que el intervalo observado encierre al verdadero par√°metro y no como la probabilidad de que el par√°metro est√© dentro del intervalo. El par√°metro no es aleatorio, los estad√≠stico $T_1$ y $T_2$ si lo son.<br><br>
* Podemos intentar construir intervalos unilaterales de la forma $(T_1, \infty)$ o $(-\infty, T_2)$ que surgen de considerar $ P(T_1 < \theta) = \gamma $ o  $ P(T_2 > \theta) = \gamma $, respectivamente.<br><br>
* La longitud del intervalo $L = T_2 - T_1$ es una variable aleatoria cuya esperanza es una medida de la precisi√≥n de nuestro intervalo. Dados dos posibles intervalos del $\gamma$ por ciento de confianza, vamos a preferir aquel que tenga menor longitud esperada.<br><br>
* A $\gamma$ lo establecemos nosotros a  priori. En general, construimos intervalos del $90$, $95$ o $99$ por ciento de confianza. La mayor√≠a de los softwares, por default, construyen al $95$ por ciento.<br><br>

‚úÖ <strong> M√©todo de la cantidad pivotal </strong><br><br>
Dada una muestra aleatoria $X_1,...,X_n$ de una poblaci√≥n $f_X(\cdot | \theta)$. Una cantidad pivotal para el par√°metro $\theta$ es una variable aleatoria $Q = q(X_1,..,X_n,\theta)$ que solo depende de la muestra y del par√°metro desconocido, y cuya distribuci√≥n es conocida.<br><br>
Al ser la distribuci√≥n de $Q$ conocida, podemos plantear:<br><br>
$$ P(q_1 < q(X_1,..,X_n,\theta) < q_2) = \gamma $$<br><br>
donde $q_1$ y $q_2$ son los cuantiles de orden $(1- \gamma)/2$ y $(1+\gamma)/2$ de la distribuci√≥n de $Q$ (esto garantiza que la longitud esperada del intervalo sea m√≠nima).<br><br>
Luego, podemos despejar $\theta$ de la siguiente forma:<br><br>
$$ P \Big( q^{-1}(X_1,..,X_n,q_1) > \theta > q^{-1}(X_1,..,X_n,q_2) \Big) = \gamma $$<br><br>
donde $q^{-1}$ es la relaci√≥n inversa de $q$ y las expresiones $q^{-1}(X_1,..,X_n,q_2)$ y $q^{-1}(X_1,..,X_n,q_2)$ representan los estad√≠sticos $T_1$ y $T_2$ buscados, respectivamente.<br><br>
üëÄ <strong>Ejemplo</strong>:<br><br>
Sea $X_1,...,X_n$ una muestra aleatoria de una poblaci√≥n normal con media y varianza desconocidas. ¬øC√≥mo podemos construir un intervalo de confianza para la media?<br><br>
Tenemos un resultado importante que nos puede ayudar a construir un intervalo:<br><br>
$$Z = \frac{\bar{X} - \mu_X}{\sqrt{\frac{\sigma^2}{n}}} \sim N(0,1)$$<br><br>
Como $Z$ depende de $\mu_X$ (la media) y de $\sigma^2$, no nos sirve como cantidad pivotal. Pero podemos usar un estimador de $\sigma^2$ para obtener una V.A. que solo depende de la muestra y del par√°metro $\mu_X$:<br><br>
$$T = \frac{\bar{X} - \mu_X}{\sqrt{\frac{S_X^2}{n}}} \sim t(n-1)$$<br><br>
La nueva variable $T$ tiene distribuci√≥n conocida (t de student) y solo depende de $\mu_X$ y de la muestra. Por lo tanto, nos sirve como cantidad pivotal para $\mu_X$. Vamos a construir un intervalo del $95 \%$ de confianza para $\mu_X$:<br><br>
$$ P \Bigg( t_1 < \frac{\bar{X} - \mu_X}{\sqrt{\frac{S_X^2}{n}}} < t_2 \Bigg) = 0.95 $$<br><br>
Esta expresi√≥n nos sirve para darnos cuenta que $t_1$ y $t_2$ deben ser los cuantiles de orden $0.25$ y $0.975$ de una t de estudent con $n-1$ grados de libertad, respectivamente. Podemos escribir:<br><br>
$$t_1 < \frac{\bar{X} - \mu_X}{\sqrt{\frac{S_X^2}{n}}} < t_2$$<br><br>
Y despejando $\mu_X$ tenemos que:<br><br>
$$ \bar{X} - t_{0.025} \frac{S_X}{\sqrt{n}} > \mu_X > \bar{X} - t_{0.975} \frac{S_X}{\sqrt{n}}$$.
De esta forma, hemos encontrado el intervalo deseado.<br><br>
Cuidado: la distribuci√≥n $t$ es sim√©trica respecto al $0$. Por lo tanto, se cumple que $t_{0.025} = - t_{0.975}$ cualesquiera sean los grados de libertad considerados.<br><br>
Cuando la poblaci√≥n es normal, son conocidas las cantidades pivotales que debemos utilizar para estimar la media y la varianza (ver intervalo de confianza para la varianza en base a una muestra proveniente de una poblaci√≥n normal).<br><br>
Cuando no conocemos la distribuci√≥n de la muestra, podemos utilizar el Teorema central del l√≠mite.<br><br>
$$T = \frac{\bar{X} - \mu_X}{\sqrt{\frac{S_X^2}{n}}} \to N(0,1)$$<br><br>
siempre que $n$ sea grande (como m√≠nimo mayor a 30).


</p>

```python
#Intervalo del 95% confianza para la media de una distribuci√≥n desconocida
muestra = np.random.poisson(2, size=30) #Recuerde que la media de una poisson es su par√°metro

q_025 = sc.stats.norm.ppf(0.025, 0, 1)
q_975 = sc.stats.norm.ppf(0.975, 0, 1)

(np.mean(muestra) - q_975 * np.sqrt(np.var(muestra)/len(muestra)), np.mean(muestra) - q_025 * np.sqrt(np.var(muestra)/len(muestra)))
```

```python
m = 1000
media = 2
q1 = sc.stats.norm.ppf(0.025, 0, 1)
q2 = sc.stats.norm.ppf(0.975, 0, 1)

decision = []
for i in range(m):
  muestra = np.random.poisson(media, size=30)
  intervalo = (np.mean(muestra) - q2 * np.sqrt(np.var(muestra)/len(muestra)), np.mean(muestra) - q1 * np.sqrt(np.var(muestra)/len(muestra)))
  if 2>intervalo[0] and 2<intervalo[1]:
    decision.append(1)
  else:
    decision.append(0)

sum(decision)/m
```

# **<font color="DeepPink">Test de Hip√≥tesis</font>**














</p>

 <p align="justify">

Una hip√≥tesis estad√≠stica es una afirmaci√≥n acerca de la distribuci√≥n de
una o m√°s variables aleatorias. Ejemplos:<br><br>
1. La V.A. $X$ tiende distribuci√≥n Normal.<br><br>
2. La media de $X$ es menor que 3.<br><br>
3. $X$ y $Y$ presentan una correlaci√≥n positiva.<br><br>

Las hip√≥tesis se plantean de a dos:<br><br>
* Hip√≥tesis Nula ($H_0$): es la hip√≥tesis a ser probada, generalmente se
refiere a la hip√≥tesis que se quiere rechazar.<br><br>
* Hip√≥tesis Alternativa ($H_1$): Es la negaci√≥n de la hip√≥tesis nula.<br><br>

Un test de hip√≥tesis es una regla o procedimiento para decidir si se rechaza o no una Hip√≥tesis nula. En base a esta regla se pueden dar 4 situaciones:



</p>

image.png

En base a esto definimos dos probabilidades que representan el tama√±o de cada error posible:<br><br>
$$ \alpha = P(\text{Rechazar }H_0 | H_0 \text{ es verdadera}) $$<br><br>
$$ \beta = P(\text{No rechazar }H_0 | H_0 \text{ es falsa}) $$<br><br>


image.png

<p align="justify">
Lo ideal ser√≠a proponer un procedimiento que nos permita minimizar los tama√±os de los dos tipos de errores simultaneamente. Esto no es posible.<br><br>
La pr√°ctica consiste en fijar $\alpha$, el tama√±o del error tipo I, en un valor "peque√±o" y hacer "lo que se pueda" con el tama√±o de error tipo II. Se considera que cometer un error de tipo I es peor que cometer un error de tipo II. Por este motivo se determina un valor $\alpha$ peque√±o y se rechaza la $H_0$ solo si existe una fuerte evidencia en su contra.<br><br>
Si existe m√°s de un test de hip√≥tesis para testear la misma hip√≥tesis, entonces se elije aquel que presente un tama√±o de error tipo II menor.

</p>

<p align="justify">
<strong>¬øC√≥mo testeamos una hip√≥tesis?</strong><br><br>
Tenemos 3 m√©todos que veremos con un ejemplo espec√≠fico.<br><br>
‚úÖ <strong>Construir un intervalo de confianza</strong><br><br>
Supongamos que tenemos una muestra aleatoria de $X_1,..,X_n$ de una variable aleatoria y queremos determinar si su media es $0$ o no. Podemos construir un intervalo de confianza adecuado para la media y calcular el intervalo observado. Si $0$ cae dentro del intervalo observado entonces no rechazamos la hip√≥tesis nula y si $0$ queda fuera del intervalo observado entonces rechazamos la hip√≥tesis nula. En este caso el tama√±o de error tipo I, tambi√©n llamado nivel de significancia de la prueba, es equivalente al $100-\gamma$ por ciento, donde $\gamma$ es el nivel de confianza al que construimos el intervalo.

</p>

```python
#Construimos una muestra simulada de una distribuci√≥n normal con media 0.05 y varianza 1
#np.random.seed(12321)
muestra = sc.stats.norm.rvs(loc=0.05, scale=1, size=1000)

#Construimos un intervalo del 95% de confianza para la media
#Vamos a testear la hip√≥tesis nula: media = 0

q_025 = sc.stats.norm.ppf(0.025, 0, 1)
q_975 = sc.stats.norm.ppf(0.975, 0, 1)

(np.mean(muestra) - q_975 * np.sqrt(np.var(muestra)/len(muestra)), np.mean(muestra) - q_025 * np.sqrt(np.var(muestra)/len(muestra))), np.mean(muestra)
```

<p align="justify">
‚úÖ <strong>Estad√≠stico de prueba</strong><br><br>
Buscamos un estad√≠stico cuya distribuci√≥n sea conocida <strong>suponiendo</strong> que la $H_0$ es verdadera. Calculamos el valor observado del estad√≠stico y lo comparamos con el o los valores cr√≠ticos que determinan si la hip√≥tesis nula debe ser o no rechazada. Ejemplo: como en el ejemplo anterior vamos a testear:<br><br>
$H_0 : \mu = 0$ versus $H_1: \mu \neq 0$<br><br>
Sabemos que:<br><br>
$$ T=\frac{\bar{X} - \mu}{\sqrt{\frac{S^2}{n}}} \to N(0,1) $$<br><br>
Si la hip√≥tesis nula es verdadera, entonces:<br><br>
$$ T=\frac{\bar{X} - 0}{\sqrt{\frac{S^2}{n}}} \to N(0,1) $$<br><br>
Calculamos el estad√≠stico observado $t$ y lo comparamos con los valores cr√≠ticos que vienen dados por los cuantiles $z_{\alpha/2}$ y $z_{1- \alpha/2}$ de una normal estandar. <br><br>
El test consiste en rechazar $H_0$ si y solo si $t<z_{\alpha/2}$ o $t>z_{1- \alpha/2}$.<br><br>
Continuando con nuestro ejemplo:



</p>

```python
#Dada nuestra muestra:

#Calculamos el estad√≠stico observado
t = (np.mean(muestra) - 0) / np.sqrt(np.var(muestra)/len(muestra))

#Lo comparamos con los cuantiles 0.025 y 0.975 de una normal estandar
q_025 = sc.stats.norm.ppf(0.025, 0, 1)
q_975 = sc.stats.norm.ppf(0.975, 0, 1)

t, q_025, q_975
```

‚úÖ <strong> Importante</strong><br><br>
Las hip√≥tesis nulas pueden ser del tipo:<br><br>
1. $H_0: \mu \leq \mu_0$ <br>
o <br>
2. $H_0: \mu \geq \mu_0$<br><br>
En estos casos el estad√≠stico de prueba no cambia, continua siendo:<br><br>
$$ T=\frac{\bar{X} - \mu_0}{\sqrt{\frac{S^2}{n}}} \to N(0,1) $$<br><br>
Pero, para el primer caso el test es:<br><br>
Rechazar $H_0$ si y solo si $t>z_{1-\alpha}$ <br><br>
y para el segundo:<br><br>
Rechazar $H_0$ si y solo si $t<z_{\alpha}$



<p align="justify">
‚úÖ <strong>M√©todo del p-valor</strong><br><br>
Calculamos el valor observado del estad√≠stico de prueba y luego obtenemos la probabilidad de observar dicho valor a trav√©s de su distribuci√≥n conocida suponiendo hip√≥tesis nula conocida. Finalmente, comparamos esta probabilidad con el tama√±o de error tipo I, $\alpha$.<br><br>
Para nuestro caso, si testeamos:<br><br>
$H_0: \mu = \mu_0$ versus $H_1: \mu \neq \mu_0$.<br><br>
El estad√≠stico de prueba es:<br><br>
$$ T=\frac{\bar{X} - \mu_0}{\sqrt{\frac{S^2}{n}}} \to N(0,1) $$<br><br>
Computamos el valor observado del estd√≠stico de prueba $t$. Finalmente, como la prueba es a dos colas, computamos el $p$-valor:<br><br>
$$ p-valor = 2 \times \min \{ P(T<t), P(T>t) \} = 2 \times P(T>|t|) $$<br><br>
El test consiste en rechazar $H_0$ si y solo si $p$-valor es menor a $\alpha$.<br><br>
Pruebas a una cola<br><br>
Si $H_0: \mu \leq \mu_0$ entonces $p$-valor $= P(T > t)$<br><br>
Si $H_0: \mu \geq \mu_0$ entonces $p$-valor $= P(T < t)$<br><br>







</p>

```python
#Calculamos el estad√≠stico observado
t = (np.mean(muestra) - 0) / np.sqrt(np.var(muestra)/len(muestra))

2 * min(sc.stats.norm.cdf(t, loc=0, scale=1) , 1-sc.stats.norm.cdf(t, loc=0, scale=1))
```

2. Determinar el tama√±o de error tipo I, generalmente $1\%$, $5\%$, hasta $10\%$.<br><br>
3. De ser posible, elegir el test con menor tama√±o de error tipo II (m√°xima potencia del test).<br><br>

7. Rechazar $H_0$ si y solo si $p$-valor es menor al tama√±o de error tipo I.


# **<font color="DeepPink">Modelos de regresi√≥n lineal</font>**

 <p align="justify">
Hasta el momento hemos intentado estudiar como se comporta una V.A. $Y$ condicionada a que otra V.A. $X$ estudiando un cierto par√°metro espec√≠fico sin suponer cual es la relaci√≥n por la cual $X$ afecta a $Y$.<br><br>
En este punto, vamos a suponer que existe una relaci√≥n real dada por alguna funci√≥n $f$ entre las V.A. $X$ e $Y$, de forma tal que si tenemos una muestra aleatoria $(X_1, Y_1),...,(X_n, Y_n)$ entonces se cumple que:<br><br>
$$ Y_i = f(X_i) + \mu_i $$<br><br>
donde $\mu_i$ es una variable aleator√≠a (o t√©rmino aleatorio) la cual suponemos que cumple:<br><br>
$$ E(\mu_i) = 0 \quad \forall i$$<br>
$$ \text{var}(\mu_i) = \sigma^2 \quad \forall i $$<br><br>
Bajo estos supuestos, la Esperanza $Y$ condicional a que $X$ toma un valor fijo $x$ cualquiera viene dado por:<br><br>
$$ E(Y|X=x) = f(x) $$ <br><br>
Con lo cual, si conocemos la funci√≥n $f$ es facil obtener todas las esperanzas condicionadas de $Y$. <br><br>
Dado que en la pr√°ctica no conocemos la forma funcional de $f$ lo que generalmente se hace es suponerla. Y la forma funcional m√°s sencilla (y m√°s utilizada) es la forma lineal:<br><br>
$$ Y_i = \beta_0 + \beta_1X_i + \mu_i $$<br><br>
donde $\beta_0$ y $\beta_1$ son constantes que deben ser estimadas utilizando los datos de una muestra.<br><br>
Bajo esta forma funcional propuesta, los datos que observamos de una muestra $(x_1,y_1), (x_2,y_2),...,(x_n,y_n)$ se generan de la siguiente forma: <br><br>
$$ y_1 = \beta_0 + \beta_1 x_1 + \mu_1 \\ y_2 = \beta_0 + \beta_1 x_2 + \mu_2 \\ ... \\ y_n = \beta_0 + \beta_1 x_n + \mu_n $$<br><br>
Dado, que no conocemos $\mu_1, \mu_2, ... ,\mu_n$ tambi√©n debemos estimarlos junto con $\beta_1$ y $\beta_2$, y lo vamos a hacer de forma tal que el aporte de estos t√©rminos aleatorios sea el menor posible dentro del modelo.<br><br>
En el mejor de los casos los t√©rminos aleatorios ser√≠an todos 0 y tendr√≠amos:<br><br>
$$ y_1 = \beta_0 + \beta_1 x_1 \\ y_2 = \beta_0 + \beta_1 x_2  \\ ... \\ y_n = \beta_0 + \beta_1 x_n  $$<br><br>
Esto genera un conjunto de ecuaciones todas vinculadas por los mismos $\beta_0$ y $\beta_1$ al cual lo podemos escribir en forma matricial:<br><br>
$$ \begin{bmatrix} y_1 \\ y_2 \\ ... \\ y_n \end{bmatrix} = \begin{bmatrix} 1 & x_1 \\ 1 & x_2 \\ ... \\ 1 & x_n \end{bmatrix} \begin{bmatrix} \beta_0 \\ \beta_1 \end{bmatrix} $$<br><br>
En general, este sistema se escribe:<br><br>
$$ Y = X \beta $$<br><br>
donde en esta √∫tlima ecuaci√≥n $Y$ representa el vector de observaciones que deseamos explicar, $X$ la matriz de caracteristicas observadas que utilizamos para explicar a $Y$ y $\beta$ es el vector de par√°metros desconocidos a estimar (o incognitas del sistema).<br><br>
Dado que al modelo le hemos sacado un t√©rmino ($\mu_i$), el sistema anterior es inconsitente y generalmente tiene m√°s ecuaciones que incognitas. Podemos encontrar un vector $\hat{\beta}$ que mejor se aproxima a una soluci√≥n. O lo que es lo mismo, buscamos un vector $\hat{\beta}$ que resuelva el siguiente problema:<br><br>
$$ \hat{\beta} = \arg \min_{\beta} || Y - X \beta  ||_2 $$<br><br>
o, equivalentemente:<br><br>
$$ \hat{\beta} = \arg \min_{\beta} || Y - X \beta  ||^2_2 $$<br><br>
Para este caso, $|| Y - X \beta  ||^2_2$, que es funci√≥n de $\beta$, es la funci√≥n de p√©rdida antes mencionada.<br><br>
Ya sabemos que la soluci√≥n a este problema viene dada por:<br><br>
$$ \hat{\beta}=(X'X)^{-1}(X'Y) $$<br><br>
A este $\hat{\beta}$ se lo conoce como estimador de M√≠nimos Cuadrados Ordinarios.

</p>

 <p align="justify">
<strong>Ejemplo</strong><br><br>
Supongamos que el salario de las personas viene determinado por la siguiente relaci√≥n:<br><br>
$$ \text{salario} = \beta_0 + \beta_1 \text{ a√±os de estudio } + \mu $$<br><br>
Si bien no tenemos los a√±os de estudio de una persona, podemos construir una variable que los aproxime en base al m√°ximo nivel educativo alcanzado por cada sujeto encuestado. La variable "estudio" la construiremos de la siguiente forma:<br><br>
$$ \text{estudio} = \begin{cases} 0 \text{ si Educacion = 0} \\ 3 \text{ si Educacion = 1} \\ 6 \text{ si Educacion = 2} \\ 9 \text{ si Educacion = 3} \\12 \text{ si Educacion = 4} \\ 15 \text{ si Educacion = 5} \\ 18 \text{ si Educacion = 6} \end{cases} $$


</p>

<strong>Supuestos cl√°siscos y sus consecuencias</strong><br><br>
1. Modelo lineal:<br><br>
$$ Y_i = X_i'\beta + \mu_i $$<br><br>
donde $X_i$ es un vector de caracter√≠sticas y $\beta$ es un vector de par√°metros.<br><br>
2. No existe perfecta colinealidad entre las caracter√≠sticas:<br><br>
$$ \mathrm{rango}(X'X) = \text{n√∫mero de regresores} $$<br><br>
Estos dos supuestos (1 y 2) permiten aplicar MCO.<br><br>
3. $$\mathrm{cov}(X_i, \mu_i) = 0.$$ <br><br>Este supuesto, junto con el de forma funcional correcta, garantiza que los estimadores sean insezgados.<br><br>
4. Homocedasticidad y no correlaci√≥n entre los errores:<br><br>
$$ \mathrm{var}(\mu_i) = \sigma^2 \quad \forall i $$<br><br>
$$ \mathrm{cov}(\mu_i, \mu_j) = 0 \quad \forall i\neq j $$<br><br>
5. $\mu_i \sim N(0, \sigma^2)$<br><br>
Los supuestos 4 y 5 permiten realizar los test de hip√≥tesis adecuados.<br><br>

# **<font color="DeepPink">Modelo logit (o probit?)</font>**

<p align="justify">
Sea $Y$ una V.A. que toma valores en $\{ 0,1 \}$, dicot√≥mica, y sea $X$ un vector aleatorio relacionado con $Y$. El modelo logit asume:<br><br>
$$ P(Y=1 | X=x) = \pi_{\beta}(x)=\frac{\text{exp}(x.\beta)}{1 + \text{exp}(x.\beta)} $$<br><br>
donde $\text{exp}(\cdot)$ es la funci√≥n exponencial y $x.\beta$ es el producto punto entre el vector $x$ de caracteristicas condicionadas y $\beta$ el vector de par√°metros a estimar.<br><br>
Los par√°metros del modelo se deben estimar por m√°xima verosimilitud. Si tenemos las observaciones de una muestra aleatoria $(y_1, x_1), (y_2, x_2),...,(y_n, x_n)$, la funci√≥n de verosimilitud viene dada por:<br><br>
$$ L(\beta) = \prod_{i=1}^{n} \pi_{\beta}(x_i)^{y_i}(1-\pi_{\beta}(x_i))^{1-y_i} $$<br><br>
Podemos aplicar logaritmo en ambos miembros de la ecuaci√≥n para obtener:<br><br>
$$ \log(L(\beta)) = \sum_{i=1}^{n} \bigg[ y_i \log \big(\pi_{\beta}(x_i) \big) + (1-y_i) \log \big( 1-\pi_{\beta}(x_i) \big) \bigg] $$<br><br>
Debemos recordar que $Y_i$ toma valor $0$ o $1$. A la funci√≥n anterior se la conoce como log-verosimilitud.<br><br>
Escrito en forma completa, para estimar el vector $\beta$ debemos resolver el siguiente problema:<br><br>
$$ \max_{\beta}  \sum_{i=1}^{n} \bigg[ y_i \log \bigg(\frac{\text{exp}(x_i .\beta)}{1 + \text{exp}(x_i .\beta)} \bigg) + (1-y_i) \log \bigg( \frac{1}{1 + \text{exp}(x_i .\beta)} \bigg) \bigg] $$:<br><br>
Este problema solo puede ser resuelto usando m√©todos num√©ricos.<br><br>
Una vez estimado el modelo se considera la siguiente regla de predicci√≥n:<br><br>
$$ \hat{Y}= \begin{cases} 1 \text{ si } \hat{\pi}_{\beta}(x) > k \\ 0 \text{  en otro caso} \end{cases} $$<br><br>
La regla es sensilla, se asigna valor $1$ a $Y$ siestimamos que su probabilidad de tomar valor $1$ es mayor a un cierto umbral $k$. Por lo general, consideramos $k=0.5$.
</p>

```python
from scipy.optimize import minimize_scalar
```

```python
dico = []
for w in Base_limpia['P21']:
  dico.append(int(w>100000))
dico = np.array(dico)
```

```python
def logv(beta0, beta1, y, x):
  sum = 0
  for i in range(len(y)):
    if y[i] == 1:
      sum +=  np.log(np.exp(beta0 + beta1*x[i]) / (1 + np.exp(beta0 + beta1*x[i])))
    else:
      sum +=  np.log(1 / (1 + np.exp(beta0 + beta1*x[i])))
  return sum
```

```python
def opt(Beta0):
  return (-1)*logv(Beta0[0], Beta0[1], dico, np.array(Base_limpia['Estudio']))
```

```python
Beta0 = (0, 0)
Beta_hat = sc.optimize.minimize(opt, Beta0, method='BFGS').x
Beta_hat
```

```python
def prob(x, beta0, beta1):
  return np.exp(beta0 + beta1*x) / (1 + np.exp(beta0 + beta1*x))
```

```python
X = np.linspace(0,40,1000)
mitad = np.linspace(0.5,0.5,1000)
Y = []
for x in X:
  Y.append(prob(x, Beta_hat[0], Beta_hat[1]))

fig, ax = plt.subplots(figsize=(9,6))
ax.plot(X, Y, label='Logit', color='green')
ax.plot(X, mitad, color='black')
ax.legend()
plt.show
```
